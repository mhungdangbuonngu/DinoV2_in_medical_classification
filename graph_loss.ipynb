```python
import pandas as pd
data=pd.read_json("E:/GitHub/DinoV2_in_medical_classification/loss/overall_loss.json")
data
```

```python
data_array=np.array(data)
overall_loss=np.mean(data_array,axis=0)
overall_loss
```
